```python
import pandas as pd  #Importing pandas as pd
# Load dataset
df = pd.read_csv(r"C:\Users\user\Downloads\customer_personal\Customer Personality Analysis Dataset.csv")  
print(df.head(5)) #printting the top 5 rows of the data

```

```python
# Check the shape of the dataset i.e., rows, columns
print("Shape of the dataset:", df.shape)
```

```python
# Summarizing data
print("Data Information: ")
df.info()
```

```python
# Checking missing values in each column
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
#Handle missing values

#Fill missing values in Z_Revenue with median
df['Z_Revenue'].fillna(df['Z_Revenue'].median(), inplace=True)
```

```python
# Fill missing values in 'Education' with the mode (most frequent value)
df['Education'].fillna(df['Education'].mode()[0], inplace=True)
```

```python
# For 'Marital_Status', fill with mode
df['Marital_Status'].fillna(df['Marital_Status'].mode()[0], inplace=True)
```

```python
# Fill missing values for 'Kidhome', 'Teenhome', 'Recency', etc. with 0
df['Kidhome'].fillna(0, inplace=True)
df['Teenhome'].fillna(0, inplace=True)
df['Recency'].fillna(df['Recency'].median(), inplace=True)
```

```python
# For monetary columns, fill with the median
monetary_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 
                    'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 
                    'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 
                    'NumWebVisitsMonth', 'AcceptedCmp3', 'AcceptedCmp4', 
                    'AcceptedCmp5', 'AcceptedCmp1', 'AcceptedCmp2', 'Complain', 
                    'Z_CostContact']

for col in monetary_columns:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
# For 'Response', fill with mode
df['Response'].fillna(df['Response'].mode()[0], inplace=True)
```

```python
# Fill with a default date (for example., current date)
df['Dt_Customer'].fillna(pd.to_datetime('2023-10-01'), inplace=True)  # Replace with your desired date
```

```python
# Check for missing values again
print("Missing values after handling:")
print(df.isnull().sum())
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Checking the number of rows after removing duplicates
print(f"Number of rows after removing duplicates: {df.shape[0]}")
```

```python
# Standardizing text values 
#zFunction to clean text
def standardize_text(value):
    if isinstance(value, str):  # Check if it's text
        return value.lower().strip()  # Make it lowercase and remove spaces
    return value  # If not text, leave it as is

# Columns we want to standardize
columns_to_standardize = ['Education', 'Marital_Status']

# Columns we want to standardize
columns_to_standardize = ['Education', 'Marital_Status']

# Example of changing specific terms
education_mapping = {
    'phd': 'doctorate',
    'master': 'masters',
    'bachelor': 'bachelors',
    'high school': 'high school',
}

# Apply the mapping
if 'Education' in df.columns:
    df['Education'] = df['Education'].replace(education_mapping)

# Check the unique values to see the results
print("Unique values in standardized columns:\n")
for col in columns_to_standardize:
    if col in df.columns:
        print(f"{col}: {df[col].unique()}")
```

```python
# Define expected values for Education
expected_education = ['graduation', 'doctorate', 'masters', 'basic']

# Clean the Education column
df['Education'] = df['Education'].apply(lambda x: x if x in expected_education else None)

# Define a mapping for Marital_Status
marital_status_mapping = {
    'together': 'married',
    '1': 'single',
    '0': 'single',
    'absurd': None,  # Remove or replace with None
    'yolo': None,    # Remove or replace with None
    'alone': 'single',
    'widow': 'widowed',
}

# Clean the Marital_Status column
df['Marital_Status'] = df['Marital_Status'].replace(marital_status_mapping)

# Check for NaN values and handle them if necessary
df['Education'].fillna('unknown', inplace=True)  # Replace NaN with 'unknown' or another value
df['Marital_Status'].fillna('unknown', inplace=True)  # Replace NaN with 'unknown' or another value

# Check the unique values after cleaning
print("Unique values in cleaned columns:")
print("Education:", df['Education'].unique())
print("Marital_Status:", df['Marital_Status'].unique())
```

```python
#Viewing the atltered data
df.head(5)
```

```python
# Check the original format of the date column
print("Original Dt_Customer values:")
print(df['Dt_Customer'].unique())
```

```python
# Convert the 'Dt_Customer' column to datetime
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], errors='coerce')

# Format the dates to 'dd-mm-yyyy'
df['Dt_Customer'] = df['Dt_Customer'].dt.strftime('%d-%m-%Y')

# Check the unique values after formatting
print("Formatted Dt_Customer values:")
print(df['Dt_Customer'].unique())
```

```python
# Check original column names
print("Original column names:")
print(df.columns)
```

```python
# Rename columns: convert to lowercase, replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('-', '_')

# Check the new column names
print("Renamed column names:")
print(df.columns)
```

```python
# Check current data types
print("Current data types:\n")
print(df.dtypes)
```

```python
 #Check the data type of 'year_birth'
print("Data type of 'year_birth':", df['year_birth'].dtype)
```

```python
# Convert 'year_birth' to numeric, forcing errors to NaN
df['year_birth'] = pd.to_numeric(df['year_birth'], errors='coerce')

# Check for any NaN values after conversion
print("NaN values in 'year_birth':", df['year_birth'].isna().sum())
```

```python
# Calculate age if 'year_birth' is valid
if 'year_birth' in df.columns:
    current_year = pd.to_datetime('now').year
    df['age'] = current_year - df['year_birth']  # Calculate age

# Handle NaN values in 'age'
if df['age'].isna().any():
        # Option 1: Fill NaN with a default value (e.g., -1)
    df['age'].fillna(-1, inplace=True)

#otherwise we can also drop rows with NaN values 
```
